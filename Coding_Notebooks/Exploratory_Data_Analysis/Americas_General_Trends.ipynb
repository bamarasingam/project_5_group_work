```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The Freedom House Project Indicators: Civil Liberties & Political Rights

```python
civil_liberties = pd.read_csv('../../Datasets/Processed_Data/CivilLiberties_Indicator_0-60_2010-2023.csv')
civil_liberties = civil_liberties.T
```

```python
civil_liberties.reset_index(inplace=True)
civil_liberties = civil_liberties.rename(columns={'index':'Year', 
                                0:'Brazil', 
                                1:'El_Salvador', 
                                2:'Honduras',
                                3:'Mexico',
                                4:'United_States'})
```

```python
civil_liberties = civil_liberties[civil_liberties['Year'] != 'Country' ]
civil_liberties = civil_liberties.sort_values(by='Year')
```

```python
civil_liberties.reset_index(drop=True,inplace=True)
civil_liberties
```

```python
civil_liberties.plot(figsize=(10,6), 
                     x='Year', 
                     title='Civil Liberties Score', 
                     ylabel='Country', 
                     fontsize=12,
                     lw=5), 
plt.legend(loc='lower left')
plt.savefig('../../Images/Civil_Liberties.png');
```

```python
political_rights = pd.read_csv('../../Datasets/Processed_Data/PoliticalRights_Indicator_0-40_2010-2023.csv')
political_rights = political_rights.T
```

```python
political_rights.reset_index(inplace=True)
political_rights = political_rights.rename(columns={'index':'Year', 
                                0:'Brazil', 
                                1:'El_Salvador', 
                                2:'Honduras',
                                3:'Mexico',
                                4:'United_States'})
```

```python
political_rights = political_rights[political_rights['Year'] != 'Country' ]
political_rights = political_rights.sort_values(by='Year')
```

```python
political_rights.reset_index(drop=True)
political_rights
```

```python
political_rights.plot(figsize=(10,6), 
                     x='Year', 
                     title='Political Rights Score', 
                     ylabel='Country', 
                     fontsize=12,
                     lw=5), 
plt.legend(loc='lower left')
plt.savefig('../../Images/Political_Rights.png');
```

# United Nations Office on Drugs and Crime: Americas Data 

```python
UN_data = pd.read_csv('../../Datasets/Processed_Data/Americas_Incarceration_Rate_2010-2021.csv')
UN_data.head()
```

```python
UN_data.drop(columns=['Region', 'Subregion', 'Indicator', 'Source'], inplace=True)
```

```python
UN_data['Country'].unique()
```

```python
target_countries = UN_data.loc[UN_data['Country'].isin(['Brazil', 'El Salvador', 'Honduras', 'Mexico', 'United States of America'])]

# UN_data.loc[(UN_data['Country'] == 'Brazil') & (UN_data['Country'] == 'El Salvador') & (UN_data['Country'] == 'Honduras') & (UN_data['Country'] == 'Mexico') & (UN_data['Country'] == 'United States of America')]
```

```python
target_countries.tail()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=target_countries, x='Year', y='VALUE', hue='Country', lw=5)
plt.ylabel('Country')
plt.title('Incarceration per 100,000 people from 2010 to 2021')
plt.legend(loc='upper left')
plt.savefig('../../Images/Incarceration_rates_all.png');
```
